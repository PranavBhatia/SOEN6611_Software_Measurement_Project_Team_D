```python
from scipy.stats import spearmanr
import matplotlib.pyplot as plt
import pandas as pd
```

```python
project_results = pd.read_csv("Project_Results_Table_Metric_1_2_3_6.csv")
```

```python
## Branch Coverage Results 
branch_coverage_data = project_results["Branch Coverage"]
print(branch_coverage_data)
```

```python
## Mutation Scores 
mutation_score_data = project_results["Mutation score"]
print(mutation_score_data)
```

```python
## Spearman Correlation Coefficient for Branch Coverage(Metric 2) and Mutation Score(Metric 3)
print(spearmanr(branch_coverage_data, mutation_score_data))
```

```python
mutation_score_plot,branch_coverage_plot = zip(*sorted(zip(mutation_score_data,branch_coverage_data)))
```

```python
plt.plot(mutation_score_plot, branch_coverage_plot)
plt.xlabel("Mutation Score")
plt.ylabel("Branch Coverage")
plt.title("Branch Coverage vs Mutation Scores")
plt.show()
## Graph Plot for Branch Coverage(Metric 2) and Mutation Score(Metric 3)
```
