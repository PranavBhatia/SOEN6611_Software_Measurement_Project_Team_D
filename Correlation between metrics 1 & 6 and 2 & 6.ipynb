```python
from scipy.stats import spearmanr
import matplotlib.pyplot as plt
import pandas as pd
```

```python
project_results = pd.read_csv("Project_Results_Table_Metric_1_2_3_6.csv")
```

```python
## Statement Coverage Results 
statement_coverage_data = project_results["Statement Coverage"]
print(statement_coverage_data)
```

```python
## Branch Coverage Results 
branch_coverage_data = project_results["Branch Coverage"]
print(branch_coverage_data)
```

```python
## Defect Density Results 
defect_density_data = project_results["Software Defect Density"]
print(defect_density_data)
```

```python
## Spearman Correlation Coefficient for Statement Coverage(Metric 1) and Software Defect Density(Metric 6)
print(spearmanr(statement_coverage_data, defect_density_data))
```

```python
defect_density_plot,statement_coverage_plot = zip(*sorted(zip(defect_density_data,statement_coverage_data)))
```

```python
plt.plot(defect_density_plot, statement_coverage_plot)
plt.xlabel("Software Defect Density")
plt.ylabel("Statement Coverage")
plt.title("Statement Coverage vs Software Defect Density")
plt.show()
## Graph Plot for Statement Coverage(Metric 1) and Software Defect Density(Metric 6)
```

```python
## Spearman Correlation Coefficient for Branch Coverage(Metric 2) and Software Defect Density(Metric 6)
print(spearmanr(branch_coverage_data, defect_density_data))
```

```python
defect_density_plot,branch_coverage_plot = zip(*sorted(zip(defect_density_data,branch_coverage_data)))
```

```python
plt.plot(defect_density_plot,branch_coverage_plot)
plt.xlabel("Software Defect Density")
plt.ylabel("Branch Coverage")
plt.title("Branch Coverage vs Software Defect Density")
plt.show()
## Graph Plot for Branch Coverage(Metric 1) and Software Defect Density(Metric 6)
```
