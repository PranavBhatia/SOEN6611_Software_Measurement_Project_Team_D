```python
from scipy.stats import spearmanr
import matplotlib.pyplot as plt
import pandas as pd
```

```python
project_results = pd.read_csv("Project_Results_Table_Metric_1_2_3_6.csv")
```

```python
## Statement Coverage Results 
statement_coverage_data = project_results["Statement Coverage"]
print(statement_coverage_data)
```

```python
## Mutation Scores 
mutation_score_data = project_results["Mutation score"]
print(mutation_score_data)
```

```python
## Spearman Correlation Coefficient for Statement Coverage(Metric 1) and Mutation Score(Metric 3)
print(spearmanr(statement_coverage_data, mutation_score_data))
```

```python
mutation_score_plot,statement_coverage_plot = zip(*sorted(zip(mutation_score_data,statement_coverage_data)))
```

```python
plt.plot(mutation_score_plot, statement_coverage_plot)
plt.xlabel("Mutation Score")
plt.ylabel("Statement Coverage")
plt.title("Statement Coverage vs Mutation Scores")
plt.show()
## Graph Plot for Statement Coverage(Metric 1) and Mutation Score(Metric 3)
```
