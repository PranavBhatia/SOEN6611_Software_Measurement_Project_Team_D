```python
from scipy.stats import spearmanr
import matplotlib.pyplot as plt
import pandas as pd
```

```python
######################################################
############ Apache Commons Collections ##############
######################################################
results = pd.read_csv("apache_commons_collections_4.2-Coverage and Complexity-JACOCO Report.csv")
#Statement Coverage Results
statement_coverage_data = results["Statement Coverage"]
#Branch Coverage Results
branch_coverage_data = results["Branch Coverage"]
#Cyclomatic Complexity Results
cyclomatic_complexity_data = results["Total Complexity"]
```

```python
## Spearman Correlation Coefficient for Statement Coverage(Metric 1) and Cyclomatic Complexity(Metric 4)
print(spearmanr(cyclomatic_complexity_data,statement_coverage_data))
```

```python
## Spearman Correlation Coefficient for Branch Coverage(Metric 2) and Cyclomatic Complexity(Metric 4)
print(spearmanr(cyclomatic_complexity_data,branch_coverage_data))
```

```python
cyclomatic_complexity_plot,statement_coverage_plot = zip(*sorted(zip(cyclomatic_complexity_data,statement_coverage_data)))
```

```python
plt.scatter(cyclomatic_complexity_plot, statement_coverage_plot)
plt.ylabel("Statement Coverage Results")
plt.xlabel("Cyclomatic Complexity Results")
plt.title("Cyclomatic Complexity vs Statement Coverage \n (Apache Commons Collections)")
plt.show()
```

```python
cyclomatic_complexity_plot,branch_coverage_plot = zip(*sorted(zip(cyclomatic_complexity_data,branch_coverage_data)))
```

```python
plt.scatter(cyclomatic_complexity_plot, branch_coverage_plot)
plt.ylabel("Branch Coverage Results")
plt.xlabel("Cyclomatic Complexity Results")
plt.title("Cyclomatic Complexity vs Branch Coverage \n (Apache Commons Collections)")
plt.show()
```
