```python
from scipy.stats import spearmanr
import matplotlib.pyplot as plt
import pandas as pd
```

```python
######################################################
############ Apache Commons Collections ##############
######################################################
jfreechart_results = pd.read_csv("apache_commons_collections_Metric_5_6.csv")
#Software Defect Density Results 
sw_defect_density_data = jfreechart_results["Software Defect Density(Metric 6)"]
#Class Change Proneness Results
avg_class_change_proneness_data = jfreechart_results["Average Class Change Proneness(Metric 5)"]
```

```python
## Spearman Correlation Coefficient for Class Change Proneness(Metric 5) and Software Defect Density(Metric 6)
print(spearmanr(sw_defect_density_data,avg_class_change_proneness_data))
```

```python
defect_density_plot,avg_class_change_proneness_plot = zip(*sorted(zip(sw_defect_density_data,avg_class_change_proneness_data)))
```

```python
plt.plot(defect_density_plot, avg_class_change_proneness_plot)
plt.xlabel("Software Defect Density")
plt.ylabel("Average Class Change Proneness")
plt.title("Software Defect Density vs Average Class Change Proneness \n(Apache Commons Collections)")
plt.show()
## Graph Plot for Class Change Proneness(Metric 5) and Software Defect Density(Metric 6)
```
